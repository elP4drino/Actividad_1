```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#cargar archivo csv desde equipo
data = pd.read_csv("infoSinNulos.csv")
data.head()
```

```python
#Corroborar que no haya nulos
nulos = data.isnull().sum()
nulos
```

```python
#copia
data1 = data.copy()
```

##Columna 1: salon_ventas

###Histograma



```python
#Realizamos histograma de variables cuantitativas
fig = plt.figure(figsize = (5,3))
plt.hist(x=data1["salon_ventas"], color = "Gray", rwidth=0.50)
plt.title("Histograma de Precio de salon_ventas")
plt.xlabel("salon_ventas")
plt.ylabel("Frecuencia")

plt.show() #Dibujamos el histograma
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize = (5,3))
plt.boxplot(data1["salon_ventas"])
plt.title("Outliers de salon_ventas")


plt.show() #Dibujamos el histograma
```

###Metodo de desviacion estandar

```python
#Metodo aplicando desviacion estandar. Encuentro los valores extremos
y = data1["salon_ventas"]
limiteSuperior = y.mean() + 3*y.std()
limiteInferior = y.mean() - 3*y.std()

print("Limite superior permitido: ", limiteSuperior)
print("Limite inferior permitido: ", limiteInferior)
```

```python
#Encontramos los Ouliers
outliers = data1[(y>limiteSuperior) | (y<limiteInferior)]
outliers
```

```python
#Obtenemos datos limpios
dataClean = data1[(y<=limiteSuperior) & (y>=limiteInferior)]
dataClean
```

```python
dataClean.info()
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize = (10,6))
plt.boxplot(dataClean["salon_ventas"])
plt.title("Outliers de salon_ventas")

plt.show() #Dibujamos el histograma
```

###Metodo de IQR (RANGO INTER QUARTILICO)

```python
#Metodo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y = data1["salon_ventas"]

percentiles25 = y.quantile(0.25) #Q1
percentiles75 = y.quantile(0.75) #Q3
iqr = percentiles75 - percentiles25

limiteSuperiorIQR = percentiles75 + 1.5*iqr
limiteInferiorIQR = percentiles25 - 1.5*iqr

print("Limite Superior Permitido: ", limiteSuperiorIQR)
print("Limite Inferior Permitido: ", limiteInferiorIQR)
```

```python
#Encontramos los Ouliers
outliersIQR = data1[(y>limiteSuperiorIQR) | (y<limiteInferiorIQR)]
outliersIQR
```

```python
#Obtenemos datos limpios
dataCleanIQR = data1[(y<=limiteSuperiorIQR) & (y>=limiteInferiorIQR)]
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize = (10,6))
plt.boxplot(dataCleanIQR["salon_ventas"])
plt.title("Outliers de salon_ventas")


plt.show() #Dibujamos el histograma
```

###Guardar en CSV el resultado
Se guarda el resultado del DataSet sin presencia de los valores atipicos

```python
dataCleanIQR.to_csv("salon_ventas.csv")
```
