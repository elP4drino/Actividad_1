###Importar librerias

```python
import pandas as pd
```

###Cargar archivo

```python
data = pd.read_excel('Inmuebles.xlsx', header=1)
data
```

##Seccion de filtros

###Filtro1. Eliminar la columna Referencia

Solo se elimina la columna de referencias del DataSet

```python
filtro1 = data.drop(columns=['Referencia'], axis=1)
filtro1
```

###Guardar en CSV


```python
# Guardar filtro en CSV
filtro1.to_csv("Filtro1.csv")
```

###Filtro2. Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00

Solo se muestran los registros que coincidan con las fechas buscadas

```python
filtro2 = data.loc[(data['Fecha_Alta'] == '2007-04-29 00:00:00') | (data['Fecha_Alta'] == '2004-01-02 00:00:00')]
filtro2
```

###Guardar en CSV


```python
filtro2.to_csv("Filtro2.csv")
```

###Filtro3. Todos los registros de “ Tipo: parking, local, industrial y piso”

Se muestran solo los registros que coincidan con los tipos buscados

```python
filtro3 = data[data['Tipo'].isin(['Parking', 'Local', 'Industrial', 'Piso'])]
filtro3
```

###Guardar en CSV

```python
filtro3.to_csv("Filtro3.csv")
```

###Filtro4. Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”

Solo los registros que coincidan con las condiciones esperadas

```python
filtro4 = data[(data['Provincia'] == 'Girona') & (data['Superficie'] > 100)]
filtro4
```

###Guardar en CSV

```python
filtro4.to_csv("Filtro4.csv")
```

###Filtro5. Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”



```python
filtro5 = data[(data['Precio Venta'] > 1000000) | (data['Vendedor'].isin(['Carmen', 'Joaquín']))]
filtro5
```

###Guardar en CSV

```python
filtro5.to_csv("Filtro5.csv")
```

###Filtro6. Todos los registros que no sean de “Provincia : Barcelona“

```python
filtro6 = data[data['Provincia'] != 'Barcelona']
filtro6
```

###Guardar en CSV

```python
filtro6.to_csv("Filtro6.csv")
```

###Filtro7. Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“

```python
filtro7 = data[~((data['Operación'] == 'Alquiler') | (data['Tipo'] == 'Oficina'))]
display(filtro7)
```

###Guardar en CSV

```python
filtro7.to_csv("Filtro7.csv")
```
