###Importar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

###Cargar archivo

```python
df = pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
df
```

###Filtro 1
Se filtran todos los registros de la columna "NOMBRE_VENDEDOR" donde el vendedor sea "DIEGO ARMANDO LEON DURAN"

```python
#Filtrar por aparaciones en columna determinada
filtro1 = df[df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN"]
filtro1
```

###Guardar en CSV


```python
filtro1.to_csv("Filtro1.csv")
```

###Filtro 2
Se filtran los registros 120,200 y 0 pero ademas recibiendo todas las columnas de estos registros

```python
#filtro por por filas y columnas por indice
filtro2 = df.iloc[[120,200,0], :]
filtro2
```

###Guardar en CSV


```python
filtro2.to_csv("Filtro2.csv")
```

###Filtro 3
Filtro todas las filas o registros, pero solo tomando las columnas de con indice entre 1 y 3, el limite superior es no inclusivo

```python
#filtro todas las filas de las columnas 1 a 3
filtro3 = df.iloc[:,1:3]
filtro3
```

###Guardar en CSV


```python
filtro3.to_csv("Filtro3.csv")
```

###Filtro 4
Primero asigno una columna como indice en mi dataframe, para posteriormente con loc filtrar todos los resultados de la columna NOMBRE_VENDEDOR donde tenga como indice los primero parametros ["F37257", "F37436"]

```python
#Asignar como indice una columna
df1 = pd.read_excel("Detalle precios y productos fabricados 2022.xlsx", index_col=0)
df1

#filtro de filas vs columnas
filtro4 = df1.loc[["F37257", "F37436"],["NOMBRE_VENDEDOR"]]
filtro4
```

###Guardar en CSV


```python
filtro4.to_csv("Filtro4.csv")
```

###Filtro 5
Esto me permite tomar solo los primero 10 registros de mi dataframe e ignorar el resto

```python
#filtro por cabecera
filtro5 = df.head(10)
filtro5
```

###Guardar en CSV


```python
filtro5.to_csv("Filtro5.csv")
```

###Filtro 6
Me permite tomar todos los registros donde el Precio unitario sea superior a 140

```python
filtro6 = df[df["PRECIO_UNITARIO"] > 140]
filtro6
```

###Guardar en CSV


```python
filtro6.to_csv("Filtro6.csv")
```

###Filtro 7
Tomo todos los registros donde el precio unitario sea inferior a 800 y ademas su costo unitario sea mayor a 100

```python
filtro7 = df[(df["PRECIO_UNITARIO"] < 800) & (df["COSTO_UNITARIO"] > 100)]
filtro7
```

###Guardar en CSV


```python
filtro7.to_csv("Filtro7.csv")
```

###Filtro 8
Tomo todos los registros donde el precio unitario sea inferior a 800 o su costo unitario sea mayor a 100

```python
filtro8 = df[(df["PRECIO_UNITARIO"] < 800) | (df["COSTO_UNITARIO"] > 100)]
filtro8
```

###Guardar en CSV



```python
filtro8.to_csv("Filtro8.csv")
```

###Filtro 9
Tomo todos los registros donde el precio unitario no sea inferior a 800 ni su costo unitario sea mayor a 100

```python
filtro9 = df[~(df["PRECIO_UNITARIO"] < 800) & ~(df["COSTO_UNITARIO"] > 100)]
filtro9
```

###Guardar en CSV

```python
filtro9.to_csv("Filtro9.csv")
```
