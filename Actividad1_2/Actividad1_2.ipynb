###Importar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

###Cargar archivo

```python
df = pd.read_excel("Datos de Facturación.xlsx")
df
```

###Filtro 1
CVE_CLPV (Valores entre 1000 y 2000)

En este caso primero se reparan algunos datos faltantes o se modifican otros para poder ser manipulados de manera adecuada.

Despues con el filtro se toman todos los registros de la columna CVE_CLPV que tengan valores entre 1000 y 2000

```python
# Limpieza de datos con replace
df.CVE_CLPV = df.CVE_CLPV.replace({'O':'0'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'6781|':'6781'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'3612C':'3612'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'      ':'0'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2385A':'2385'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2510A':'2510'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'8440Q':'8440'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'M':'0'}, regex=False)

filtro1 = df[(df["CVE_CLPV"].astype(int) >= 1000) & (df["CVE_CLPV"].astype(int) <= 2000)]
display(filtro1)
```

###Guardar en CSV


```python
filtro1.to_csv("Filtro1.csv")
```

###Filtro 2
CVE_VEND

Filtra todos los registros donde CVE_VEND no tanga como clave 4 o 5

```python
filtro2 = df[(df["CVE_VEND"] != "4") | (df["CVE_VEND"] != "5")]
filtro2
```

###Guardar en CSV


```python
filtro2.to_csv("Filtro2.csv")
```

###Filtro 3
FECHA_ENT (fechas de 13/04/2022 12:00:00)

Filtro todos los registros donde en la columna FECHA_ENT tenga como valor 2022-02-28 00:00:00

###Filtro 4
CAN_TOT (cantidades menores a 5951.7) o estatus E

Filtro que me permite obtener todos los registros donde en la columna CAN_TOT los valores sean inferiores a 59517 o su valor en la columna STATUS sea igual a E
