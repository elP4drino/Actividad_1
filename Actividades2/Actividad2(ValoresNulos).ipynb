###Importar librerias

```python
import pandas as pd
import numpy as np
```

###Cargar archivo

```python
data = pd.read_csv("Ventas_totales.csv")
data.head()
```

###Revisar la cantidad de valores nulos por columna


```python
#Identificar valores nulos por columna para saber como ir resolviendo los nulos
valoresNulos = data.isnull().sum()
valoresNulos
```

###Eliminar columnas innecesarias

En este caso elimine la utima columna porque no ofrecia algun valor para mi, ademas de no encontrarle mucho sentido



```python
#Eliminar columnas innecesarias
data = data.drop(['otros'], axis=1)
data.head()
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = data.isnull().sum()
valoresNulos
```

###Tratar los nulos en subtotal_ventas_alimentos_bebidas

Reemplazo todos los nulos por la media de los valores existentes de esa columna, ya que es una buena opcion promediar los costos de las ventas

```python
#Primer metodo de sustitucion de valores nulos
#Sustituir valores nulos con promedio o media
data2 = data.copy()
data2["subtotal_ventas_alimentos_bebidas"] = data2["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].mean(),1))
data2.head()
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios

valoresNulos = data2.isnull().sum()
valoresNulos
```

###Tratar los nulos en salon_ventas, tarjetas_debito y tarejtas_credito

Sustituyo valores nulos por valores no nulos hacia atras backward fill, llegando a esta conclusion porque son datos a los que realmente no puedo aplicar alguna medida de tendencia central, entonces opte por agregarles los valores de los registros anteriores, pues si alguna tarjeta se uso una vez es probeble que se use de nuevo para una nueva compra o transaccion

```python
#Reemplazo los nulos por el valor anterior en el registro en las siguientes 3 columnas
data2["salon_ventas"] = data2["salon_ventas"].fillna(method="bfill")
data2["tarjetas_debito"] = data2["tarjetas_debito"].fillna(method="bfill")
data2["tarjetas_credito"] = data2["tarjetas_credito"].fillna(method="bfill")
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = data2.isnull().sum()
valoresNulos
```

###Tratar los nulos en bebidas, panaderia, lacteos, carnes, verduleria_fruteria, alimentos_preparados_rotiseria

Sustituyo valores nulos por sus maximos y minimos registrados en esas columnas, esto basado en el tipo de producto especifico

```python
#Sustituir nulos por maximo de esa columna
data2["bebidas"] = data2["bebidas"].fillna(round(data2["bebidas"].max(),1))

#Sustituir nulos por minimo de esa columna
data2["panaderia"] = data2["panaderia"].fillna(round(data2["panaderia"].min(),1))

#Sustituir nulos por maximo de esa columna
data2["lacteos"] = data2["lacteos"].fillna(round(data2["lacteos"].max(),1))

#Sustituir nulos por minimo de esa columna
data2["carnes"] = data2["carnes"].fillna(round(data2["carnes"].min(),1))

#Sustituir nulos por maximo de esa columna
data2["verduleria_fruteria"] = data2["verduleria_fruteria"].fillna(round(data2["verduleria_fruteria"].max(),1))

#Sustituir nulos por minimo de esa columna
data2["alimentos_preparados_rotiseria"] = data2["alimentos_preparados_rotiseria"].fillna(round(data2["alimentos_preparados_rotiseria"].min(),1))
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = data2.isnull().sum()
valoresNulos
```

###Tratar los nulos en otros_medios y almacen

Susituyo valores nulos por valores no nulos hacia adelante "forwad fill" ("ffill"), pues tomo el valor despues de mi espacio nulo (solo para utilizar este metodo tambien)

```python
#Reemplazar los nulos por su valor siguiente
data2["otros_medios"] = data2["otros_medios"].fillna(method="ffill")
data2["almacen"] = data2["almacen"].fillna(method="ffill")
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = data2.isnull().sum()
valoresNulos
```

###Tratar los nulos en indumentaria_calzado_textiles_hogar y electronicos_articulos_hogar

Sustituir valores nulos con la mediana

```python
data2["indumentaria_calzado_textiles_hogar"] = data2["indumentaria_calzado_textiles_hogar"].fillna(round(data["indumentaria_calzado_textiles_hogar"].mean(),1))
data2["electronicos_articulos_hogar"] = data2["electronicos_articulos_hogar"].fillna(round(data["electronicos_articulos_hogar"].median(),1))
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = data2.isnull().sum()
valoresNulos
```
